For fastsimcoal2, we'll generate an SAF file for each set of BAMs, generate sfs files with realSFS, and then fold them with realSFS (we don't know the ancestral state of the alleles)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import subprocess
```

```python
for run in list(range(1,11)):
    for model in ["one_step_geneflow_one_step_size", 
                  "one_step_geneflow_two_step_size", 
                  "two_step_geneflow_two_step_size", 
                  "two_step_geneflow_one_step_size"]:
        task = "fsc"
        mem = "72"
        cpus = "18"
        inpath = "../fastsimcoal2_files/"
        templateprefix = "apalacer"
        tplprefix = model
        run = str(run)
        subprocess.run(["sbatch --mem=" + mem + "g --ntasks=" + 
             cpus + " ../" + task + ".sh " + 
             inpath + " " + templateprefix + " " + tplprefix + " " + run ], shell=True)
```
